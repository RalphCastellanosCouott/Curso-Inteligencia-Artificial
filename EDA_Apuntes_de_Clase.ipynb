```python
# prompt: Cr√©ame 2 arreglos y matriz, el primero llamado saldo, que contiene datos sint√©ticos del saldo bancario, otro arreglo llamado antiguedad que contiene datos de la antiguedad de la cuenta bancaria de 0 a 10, dos arreglos enteros, una matriz de dos columnas que contiene la latitud en la columna 0, y la longitud en la columna 1, con datos muy cercanos a Bucaramanga, es decir que varie la latitud y longitud en decimales o centecimales. El n√∫mero de datos son 20. La matriz se llama Ubicaci√≥n. Finalmente quiero que usando column stack me una las matrices y las listas creadas. Defina la semilla aleatoria en 42.

import numpy as np

# Set random seed
np.random.seed(42)

# Create saldo array
saldo = np.random.randint(1000, 100000, size=20)

# Create antiguedad array
antiguedad = np.random.randint(0, 11, size=20)

# Create Ubicaci√≥n matrix (latitude and longitude near Bucaramanga)
# Bucaramanga approximate coordinates: 7.125¬∞ N, 73.116¬∞ W
latitud = np.random.uniform(7.10, 7.15, size=20)  # Latitude variation
longitud = np.random.uniform(-73.13, -73.10, size=20)  # Longitude variation

ubicacion = np.column_stack((latitud, longitud))

# Stack arrays and matrix using column_stack
data = np.column_stack((saldo, antiguedad, ubicacion))

print("Saldo:\n", saldo)
print("\nAntiguedad:\n", antiguedad)
print("\nUbicaci√≥n:\n", ubicacion)
print("\nDatos combinados:\n", data)
```

```python
import pandas as pd

# Dataset simulado
data = {
    "ID": [1, 2, 3, 4, 5],
    "Ciudad": ["Madrid", "Barcelona", "Sevilla", "Madrid", "Valencia"],  # Nominal
    "Educaci√≥n": ["Secundaria", "Primaria", "Universitaria", "Primaria", "Universitaria"],  # Ordinal
    "G√©nero": ["F", "M", "F", "M", "F"],  # Nominal
}
df = pd.DataFrame(data)
print("Dataset original:")
print(df)
```

```python
# Usando get_dummies para One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=["Ciudad"], prefix="Ciudad")
print("\nCodificaci√≥n One-Hot para 'Ciudad':")
print(df_onehot)
```

```python
from sklearn.preprocessing import OneHotEncoder


# Configuraci√≥n del codificador
miencoder = OneHotEncoder(sparse_output=False)  # Cambiado de sparse a sparse_output
ciudad_onehot_encoded = miencoder.fit_transform(df[["Ciudad"]])

# Crear un DataFrame con los resultados
onehot_df = pd.DataFrame(ciudad_onehot_encoded, columns=miencoder.get_feature_names_out(["Ciudad"]))
print("\nCodificaci√≥n One-Hot con Scikit-learn:")
print(onehot_df)
```
